```python

from ultralytics import YOLO

# Load a pre-trained YOLOv8 model
model = YOLO('yolov8n.pt')  # You can change this to other models like 'yolov8s.pt'

# Train the model and save checkpoints after each epoch
model.train(
    data='/content/drive/MyDrive/brand/data.yaml',  # Path to your dataset configuration
    epochs=10,  # Number of epochs to train
    imgsz=640,  # Image size used for training
    batch=32,  # Batch size
    save_period=1,  # Save checkpoint after each epoch
    project='/content/drive/MyDrive/brand/runs_new1',  # Folder to save runs/checkpoints
    name='yolov8n_brand',  # Name of the training run folder
)

# Save the final model after training
model_path = '/content/drive/MyDrive/brand/runs/yolov8n_brand/best.pt'
model.export(format='torchscript', path=model_path)  # Save full model in TorchScript format
```

```python
!pip install ultralytics
from ultralytics import YOLO

# Load the trained YOLOv8 model
model = YOLO('/content/drive/MyDrive/brand/runs_new1/yolov8n_brand/weights/best.pt')

# Inference on a single image
#results = model('/content/drive/MyDrive/joy1.jpg')
results = model('/content/drive/MyDrive/brand/pepsi.jpg', conf=0.25)  # Adjust the confidence threshold


# Display predictions on the image(s)
for result in results:
    result.show()

# Optional: Save the result image(s)
for idx, result in enumerate(results):
    # Construct a filename for saving
    save_path = f'/content/drive/MyDrive/brand/inference_results/inference_result_{idx}.jpg'
    result.save(save_path)
```

```python
import os

def count_images_in_directory(directory):
    # List of common image file extensions
    image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff', '.webp')

    # Count images by checking their extensions
    count = sum(1 for file in os.listdir(directory)
                if file.lower().endswith(image_extensions))

    print(f'Total images in "{directory}": {count}')
    return count

# Example usage
directory_path = '/content/drive/MyDrive/brand/valid/images'
count_images_in_directory(directory_path)
```
